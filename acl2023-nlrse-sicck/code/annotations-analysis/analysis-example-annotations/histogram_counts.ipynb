```python
import pandas as pd
import numpy as np
```

```python
filename = "/content/gdrive/MyDrive/Colab Notebooks/natural-logic/final-datasets/modifiers-sentences-sick-dataset.xlsx"
```

```python
df = pd.read_excel(filename, sheet_name="all-modified")
df.head()
```

```python
df.columns
```

```python
val_counts = df[['Modifier', 'Premise/Hypothesis/Both','Part of Premise/Hypothesis Modified']].value_counts().reset_index()
val_counts = val_counts.rename(columns={0:"Count"})
```

```python
mod_counts = df['Modifier'].value_counts().reset_index()
mod_counts = mod_counts.rename(columns={"index":"Modifier", "Modifier":"Count"})
# mod_counts[mod_counts['Modifier'] == "NONE"].loc[1,1] #= "No Modifier"
mod_counts.loc[mod_counts['Modifier'] == "NONE",['Modifier']] = "No Modifier"
mod_counts
```

```python
!pip install altair
```

```python
import altair as alt

bars = alt.Chart(mod_counts).mark_bar().encode(
    y=alt.Y("Count:Q"), x=alt.X("Modifier:O", axis=alt.Axis(labelAngle=-45), sort='-y'))

text = bars.mark_text(
    align='left',
    baseline='middle',
    dx=-6,  # Nudges text to right so it doesn't appear on top of the bar
    dy = -3
).encode(
    text='Count:Q'
)

(bars + text).properties(height=500, width=800).properties(
    title='Modifier-wise counts'
).configure_title(
    fontSize=20,
    font='Tahoma',
    anchor='middle',
    color='gray'
)
```

```python

bars = alt.Chart(mod_counts).mark_bar().encode(
    y=alt.Y("Count:Q"), x=alt.X("Modifier:O", axis=alt.Axis(labelAngle=-45), sort='-y'))

text = bars.mark_text(
    align='left',
    baseline='middle',
    dx=-6,  # Nudges text to right so it doesn't appear on top of the bar
    dy = -3
).encode(
    text='Count:Q'
)

(bars + text).properties(height=500, width=800).properties(
    title='Modifier-wise counts'
).configure_title(
    fontSize=20,
    font='Tahoma',
    anchor='middle',
    color='gray'
)
```

```python
mod2_counts = df['Premise/Hypothesis/Both'].value_counts().reset_index()
mod2_counts = mod2_counts.rename(columns={"index":"Premise_Hypothesis_Both", "Premise/Hypothesis/Both":"Count"})
mod2_counts.loc[mod2_counts['Premise_Hypothesis_Both'] == "NONE",'Premise_Hypothesis_Both'] = "Original"
mod2_counts
```

```python

bars = alt.Chart(mod2_counts).mark_bar().encode(
    y=alt.Y("Count:Q",scale=alt.Scale(domain=[0, 650])), x=alt.X("Premise_Hypothesis_Both:O", axis=alt.Axis(labelAngle=-45), sort='-y', title="Modification to Premise or Hypothesis or both"))

text = bars.mark_text(
    align='left',
    baseline='middle',
    dx=-6,  # Nudges text to right so it doesn't appear on top of the bar
    dy = -3
).encode(
    text='Count:Q'
)

(bars + text).properties(height=300, width=300).properties(
    title='Premise, Hypothesis Modification-wise counts'
).configure_title(
    fontSize=20,
    font='Tahoma',
    anchor='start',
    color='gray'
)
```

```python
mod3_counts = df['Part of Premise/Hypothesis Modified'].value_counts().reset_index()
mod3_counts = mod3_counts.rename(columns={"index":"Part of Sentence Modified", "Part of Premise/Hypothesis Modified":"Count"})
mod3_counts.loc[mod3_counts['Part of Sentence Modified'] == "NONE",'Part of Sentence Modified'] = "Original"
mod3_counts
```

```python
bars = alt.Chart(mod3_counts).mark_bar().encode(
    y=alt.Y("Count:Q",scale=alt.Scale(domain=[0, 820])), x=alt.X("Part of Sentence Modified:O", axis=alt.Axis(labelAngle=-45), sort='-y', title="Part of Sentence Modified"))

text = bars.mark_text(
    align='left',
    baseline='middle',
    dx=-6,  # Nudges text to right so it doesn't appear on top of the bar
    dy = -3
).encode(
    text='Count:Q'
)

(bars + text).properties(height=300, width=300).properties(
    title='Counts by part of sentence modified: Subject, Verb, Object'
).configure_title(
    fontSize=20,
    font='Tahoma',
    anchor='start',
    color='gray'
)
```
